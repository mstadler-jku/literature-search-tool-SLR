# Data Cleaning 

## Load Data

```python
import os.path
import pandas as pd
import numpy as np

# root
csv_folder_str = 'csv_files_v2'

# paths
IEEE_CSV_PATH = os.path.join(csv_folder_str, "01_ieee.csv")
ACM_CSV_PATH = os.path.join(csv_folder_str, "02_acm.csv")
SCOPUS_CSV_PATH = os.path.join(csv_folder_str, "03_scopus.csv")
WOS_CSV_PATH = os.path.join(csv_folder_str, "04_wos.csv")
SPRINGER_CSV_PATH = os.path.join(csv_folder_str, "05_springer.csv")

# read data
df_ieee = pd.read_csv(IEEE_CSV_PATH)
df_acm = pd.read_csv(ACM_CSV_PATH)
df_scopus = pd.read_csv(SCOPUS_CSV_PATH)
df_wos = pd.read_csv(WOS_CSV_PATH)
df_springer = pd.read_csv(SPRINGER_CSV_PATH)

# print length
print(f'Results in IEEE\t\t{len(df_ieee)}')
print(f'Results in ACM\t\t{len(df_acm)}')
print(f'Results in Scopus\t{len(df_scopus)}')
print(f'Results in WoS\t\t{len(df_wos)}')
print(f'Results in Springer\t{len(df_springer)}')
print('-------------------------------')
print(f'SUM\t\t\t\t\t{len(df_ieee) + len(df_acm) + len(df_scopus) + len(df_wos) + len(df_springer)}')


```

## Prep IEEE dataset

```python
df_ieee
```

```python
# df_ieee.groupby(by=["Publication Title"]).size()
df_ieee[df_ieee['Publication Title'] == "IEEE Transactions on Information Forensics and Security"]
# df_ieee[df_ieee['Publication Title'].str.contains('Conference on Software Engineering')]
```

```python
# renaming
df_ieee_prep = df_ieee.copy()
df_ieee_prep['source'] = 'IEEE Xplore'
df_ieee_prep = df_ieee_prep.rename(
    columns={'Document Title': 'title', 'Publication Year': 'pub_year', 'DOI': 'doi', 'PDF Link': 'link',
             'Document Identifier': 'type', 'Authors': 'authors', 'Abstract': 'abstract',
             'Publication Title': 'pub_title'})
# select
df_ieee_prep = df_ieee_prep[['title', 'authors', 'doi', 'link', 'pub_year', 'type', 'abstract', 'source', 'pub_title']]
df_ieee_prep = df_ieee_prep[df_ieee_prep.doi != 'DOI']  # get rid of additional header
df_ieee_prep
```

## Prep ACM dataset

```python
df_acm
```

```python
# renaming
df_acm_prep = df_acm.copy()
df_acm_prep['source'] = 'ACM Digital Library'
df_acm_prep = df_acm_prep.rename(
    columns={'Item type': 'type', 'Title': 'title', 'DOI': 'doi', 'Publication year': 'pub_year',
             'URLs': 'link', 'Authors': 'authors', 'Abstract': 'abstract', 'Proceedings title': 'pub_title'})

# fill journal names into pub_title
df_acm_prep.pub_title.fillna(df_acm_prep['Journal'], inplace=True)

# select
df_acm_prep = df_acm_prep[['title', 'authors', 'doi', 'link', 'pub_year', 'type', 'abstract', 'source', 'pub_title']]
df_acm_prep
```

## Prep Scopus dataset

```python
df_scopus
```

```python
# df_scopus.loc[df_scopus['Language of Original Document'] != 'English']
```

```python
# df_scopus.groupby(by="Source title").size()
df_scopus[df_scopus['Source title'] == 'Computers and Security']
# df_scopus[df_scopus['Source title'] == 'IEEE Transactions on Information Forensics and Security']


# IEEE Transactions on Information Forensics and Security

```

```python
# renaming
df_scopus_prep = df_scopus.copy()

# drop non-english papers
df_scopus_prep = df_scopus_prep[df_scopus['Language of Original Document'] == 'English']

# beautify Source title col
# df_scopus_prep['Source title'] = df_scopus_prep['Source title'].str[df_scopus_prep[].str.index('/'):]
# df_scopus_prep['slash_idx'] = df_scopus_prep[df_scopus_prep['Source title'].str.contains('/', na=False)]['Source title'].str.index('/')
# df_scopus_prep[df_scopus_prep['Source title'].str.contains('/', na=False)]

df_scopus_prep['source'] = 'Scopus'
df_scopus_prep = df_scopus_prep.rename(
    columns={'Document Type': 'type', 'Title': 'title', 'DOI': 'doi', 'Year': 'pub_year', 'Link': 'link',
             'Author full names': 'authors', 'Abstract': 'abstract', 'Source title': 'pub_title'})

df_scopus_prep = df_scopus_prep[df_scopus_prep.type != 'Conference review']

# select
df_scopus_prep = df_scopus_prep[
    ['title', 'authors', 'doi', 'link', 'pub_year', 'type', 'abstract', 'source', 'pub_title']]
df_scopus_prep
```

## Prep Web of Science dataset

```python
df_wos
```

```python
# renaming
df_wos_prep = df_wos.copy()

# fill journal names into pub_title
df_wos_prep['Proceedings title'].fillna(df_wos_prep['Journal'], inplace=True)

# only english papers
df_wos_prep = df_wos_prep[df_wos_prep.Language == 'English']
df_wos_prep['source'] = 'Web of Science'
df_wos_prep = df_wos_prep.rename(
    columns={'Item type': 'type', 'Title': 'title', 'DOI': 'doi', 'Publication year': 'pub_year', 'URLs': 'link',
             'Authors': 'authors', 'Abstract': 'abstract', 'Proceedings title': 'pub_title'})

# select
df_wos_prep = df_wos_prep[['title', 'authors', 'doi', 'link', 'pub_year', 'type', 'abstract', 'source', 'pub_title']]
df_wos_prep
```

## Prep SpringerLink dataset

```python
df_springer
```

```python
AND_DELIMITER = ' and '
df_springer_prep = df_springer.copy()

# drop non-english papers
df_springer_prep = df_springer_prep[df_springer_prep.language == 'en']
df_springer_prep['source'] = 'SpringerLink'
df_springer_prep['authors'] = df_springer_prep['creators/0/creator'] + AND_DELIMITER + df_springer_prep[
    'creators/1/creator'] + AND_DELIMITER + df_springer_prep['creators/2/creator'] + AND_DELIMITER + df_springer_prep[
                                  'creators/3/creator'] + AND_DELIMITER + df_springer_prep['creators/4/creator']
df_springer_prep = df_springer_prep.rename(
    columns={'contentType': 'type', 'publicationDate': 'pub_year', 'url/0/value': 'link',
             'publicationName': 'pub_title'})
# select
df_springer_prep = df_springer_prep[
    ['title', 'authors', 'doi', 'link', 'pub_year', 'type', 'abstract', 'source', 'pub_title']]
df_springer_prep
```

## Concat

```python
df_all = pd.concat([df_ieee_prep, df_acm_prep, df_scopus_prep, df_wos_prep, df_springer_prep])
df_all
```

## Remove duplicates

### Check Nulls

```python
# check title
print(df_all.title.isna().sum())
# check pub_year
print(df_all.pub_year.isna().sum())
```

### Remove duplicates by DOI

```python
# safe doi null values to add them alter again elsewhise they'd be recognized as duplicates
df_all_prep = df_all.copy()
df_only_doi_nulls = df_all[df_all.doi.isna()]
df_only_doi_nulls
```

```python
df_all_prep = df_all.dropna(subset='doi')
df_all_prep
```

```python
df_all_prep = df_all_prep.drop_duplicates(subset='doi')
df_all_prep
```

```python
## add the null values again
df_all_prep = pd.concat([df_all_prep, df_only_doi_nulls])
df_all_prep
```

### Remove duplicates by title

```python
df_all_prep = df_all_prep.drop_duplicates(subset='title')
df_all_prep
```

### Remove Technical Report

```python
df_all_prep = df_all_prep[df_all_prep.type != 'Report']
df_all_prep
```

## Export

```python
# export
df_all_prep.to_csv((os.path.join('outputs', 'cleaning_export', 'raw_literature_parser_result.csv')), sep=',', encoding='utf-8', index=True, index_label='id')
```
