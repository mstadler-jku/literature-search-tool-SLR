```python
import os.path
import pandas as pd

# root

# paths
KW_CSV_PATH = os.path.join('csv_files_v2', 'all_results_cleaned.csv')

# read data
df_kw_search = pd.read_csv(KW_CSV_PATH)

# print length
print(f'Length of dataset: \t{len(df_kw_search)}')


```

```python
# remove unwanted types
df_kw_search['type'].value_counts()
unwanted_types = ['Standards', 'Editorial', 'Short survey', 'IEEE Standards', 'Review', 'Erratum',
                  'Note', 'Data paper']
df_kw_search = df_kw_search[df_kw_search.type.isin(unwanted_types) == False]
print(len(df_kw_search))
df_kw_search
```

```python
# remove entries that are no actual papers
dois_to_remove = ['10.1109/HotSWUp.2012.6226606', '10.1109/SER-IP.2017..16', '10.1109/ICHB.2011.6094345',
                  '10.1109/MMIT.2010.1', '10.1109/AVSS.2010.3']
titles_to_remove = [
    "SIN '13: Proceedings of the 6th International Conference on Security of Information and Networks",
    "ANC '13: Proceedings of the Second ACM MobiHoc Workshop on Airborne Networks and Communications",
    "Airborne '12: Proceedings of the First ACM MobiHoc Workshop on Airborne Networks and Communications",
    "7th International Conference on Applied Computing and Information Technology, ACIT 2019"]
df_kw_search = df_kw_search[df_kw_search.doi.isin(dois_to_remove) == False]
df_kw_search = df_kw_search[df_kw_search.title.isin(titles_to_remove) == False]

# construct proper index column
df_kw_search = df_kw_search.reset_index()
df_kw_search.id = df_kw_search.index + 1
df_kw_search = df_kw_search.drop('index', axis=1)
df_kw_search
```
