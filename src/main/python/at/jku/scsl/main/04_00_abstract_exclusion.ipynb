# Load Data 

```python
from src.main.python.at.jku.scsl.util.util import concat_path, get_project_root_directory
import pandas as pd

ms_df = pd.read_csv(concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results',
                                'MS_Pre_Exclusion_abstracts_All_Paper.csv'), sep=';')
ms_df = ms_df.rename(columns={'relevant?': 'relevant_ms'})

sam_df = pd.read_csv(concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results',
                                 '01_rnd_abstract_SAM.csv'))
sam_df = sam_df.rename(columns={'relevant': 'relevant_sam'})

mike_df = pd.read_csv(concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results',
                                  '02_rnd_abstract_MIKE.csv'))
mike_df = mike_df.rename(columns={'relevant': 'relevant_mike'})

mv_df = pd.read_csv(concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results',
                                '03_rnd_abstract_MV.csv'))
mv_df = mv_df.rename(columns={'relevant': 'relevant_mv'})

```

 # Sam

```python
# compare sam
merged_df_sam = pd.merge(ms_df, sam_df, on='id', suffixes=('_ms', '_sam'))

disagreements_sam = merged_df_sam[(merged_df_sam['relevant_ms'] == 'n') & (merged_df_sam['relevant_sam'] == 'y')]
percentage_disagreement = (len(disagreements_sam) / len(merged_df_sam)) * 100

# Reorder columns to make 'relevant_ms' and 'relevant_sam' second and third
cols = disagreements_sam.columns.tolist()
# Move 'relevant_ms' and 'relevant_sam' to the second and third positions
# Assuming 'id' should be the first column
new_order = [cols[0], 'relevant_ms', 'relevant_sam'] + [col for col in cols if
                                                        col not in ['id', 'relevant_ms', 'relevant_sam']]
disagreements_sam = disagreements_sam[new_order]

print(f'Percentage of Disagreements: {percentage_disagreement:.2f}%')
# Display the disagreements and the percentage
disagreements_sam
```

 # Mike

```python
# compare mike
merged_df_mike = pd.merge(ms_df, mike_df, on='id', suffixes=('_ms', '_mike'))

disagreements_mike = merged_df_mike[(merged_df_mike['relevant_ms'] == 'n') & (merged_df_mike['relevant_mike'] == 'y')]
percentage_disagreement = (len(disagreements_mike) / len(merged_df_mike)) * 100

# Reorder columns to make 'relevant_ms' and 'relevant_sam' second and third
cols = disagreements_mike.columns.tolist()
# Move 'relevant_ms' and 'relevant_sam' to the second and third positions
# Assuming 'id' should be the first column
new_order = [cols[0], 'relevant_ms', 'relevant_mike'] + [col for col in cols if
                                                         col not in ['id', 'relevant_ms', 'relevant_mike']]
disagreements_mike = disagreements_mike[new_order]

# Display the disagreements and the percentage
print(f'Percentage of Disagreements: {percentage_disagreement:.2f}%')
disagreements_mike
```

# MV

```python
# compare mike
merged_df_mv = pd.merge(ms_df, mv_df, on='id', suffixes=('_ms', '_mv'))
merged_df_mv['relevant_mv'] = merged_df_mv['relevant_mv'].str.strip().str.replace(r'\s+', ' ', regex=True).str.lower()

disagreements_mv = merged_df_mv[(merged_df_mv['relevant_ms'] == 'n') & (merged_df_mv['relevant_mv'] == 'y')]
percentage_disagreement = (len(disagreements_mv) / len(merged_df_mv)) * 100

# Reorder columns to make 'relevant_ms' and 'relevant_sam' second and third
cols = disagreements_mv.columns.tolist()
# Move 'relevant_ms' and 'relevant_sam' to the second and third positions
# Assuming 'id' should be the first column
new_order = [cols[0], 'relevant_ms', 'relevant_mv'] + [col for col in cols if
                                                       col not in ['id', 'relevant_ms', 'relevant_mv']]
disagreements_mv = disagreements_mv[new_order]

# Display the disagreements and the percentage
print(f'Percentage of Disagreements: {percentage_disagreement:.2f}%')
disagreements_mv
```

```python
all_disagreements = pd.concat([disagreements_mv, disagreements_mike, disagreements_sam])
all_disagreements
```

```python
df_disagreement_resolve_mv = pd.read_csv(
    concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results', 'disagreements',
                'Abstract_Second_Opinion_MV - Sheet1.csv'))
df_disagreement_resolve_mike = pd.read_csv(
    concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results', 'disagreements',
                'Abstract_Second_Opinion_Mike - Sheet1.csv'))
df_disagreement_resolve_sam = pd.read_csv(
    concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results', 'disagreements',
                'Abstract_Second_Opinion_Sam - Sheet1.csv'))


df_all_disagreement_resolve = pd.concat([df_disagreement_resolve_mv, df_disagreement_resolve_mike, df_disagreement_resolve_sam])
df_all_disagreement_resolve
```

```python

merged_disagreement_resolve = pd.merge(df_all_disagreement_resolve, all_disagreements, on='id')
merged_disagreement_resolve
```

```python
# Desired columns order
desired_columns = ['id', 'relevant_ms', 'relevant_MV_disagree', 'relevant_MIKE_disagree', 'relevant_SAM_disagree']

# Add the remaining columns to the desired order list
remaining_columns = [col for col in merged_disagreement_resolve.columns if col not in desired_columns]
new_column_order = desired_columns + remaining_columns

# 'A lightweight, adaptive security mechanism' is twice
merged_disagreement_resolve = merged_disagreement_resolve.drop_duplicates(subset=['id'])

# Reorder the DataFrame columns
merged_disagreement_resolve = merged_disagreement_resolve[new_column_order]

merged_disagreement_resolve['relevant_MV_disagree'] = merged_disagreement_resolve[
    'relevant_MV_disagree'].str.strip().str.lower()
merged_disagreement_resolve
```

```python


# Function to check if non-NaN values in both columns are different
def keep_row(row):
    ms_value = row['relevant_ms']
    mv_value = row['relevant_MV_disagree']
    mike_value = row['relevant_MIKE_disagree']
    sam_value = row['relevant_SAM_disagree']
    if not pd.isna(mv_value):
        return ms_value != mv_value
    elif not pd.isna(mike_value):
        return ms_value != mike_value
    elif not pd.isna(sam_value):
        return ms_value != sam_value
    print(f'WARNING: row has no match for ID {row["id"]}')
    return True


# Apply the function and filter the DataFrame
filtered_df = merged_disagreement_resolve[merged_disagreement_resolve.apply(keep_row, axis=1)]

filtered_df
```
