## Get full abstract 

```python
df_abs_data = pd.read_csv(os.path.join('outputs', 'cleaning_export', 'cleaning_p2.csv'))
df_abs_data.sort_values('id')
df_manual_parse.sort_values('id')

# insert non-abbreviated data
df_manual_parse.abstract = df_abs_data.abstract
df_manual_parse.pub_title = df_abs_data.pub_title
df_manual_parse.authors = df_abs_data.authors
df_manual_parse.title = df_abs_data.title

# get only the ones voted yes
df_manual_parse_vote_yes = df_manual_parse[df_manual_parse['relevant?'] == 'y']

# rename old identifier
df_manual_parse_vote_yes = df_manual_parse_vote_yes.rename(columns={'id': 'id_title'})

# construct new identifier
df_manual_parse_vote_yes = df_manual_parse_vote_yes.reset_index()
df_manual_parse_vote_yes['id'] = df_manual_parse_vote_yes.index + 1
df_manual_parse_vote_yes = df_manual_parse_vote_yes.drop('index', axis=1)
df_manual_parse_vote_yes
```

```python
df_manual_parse_vote_yes.to_csv(
    os.path.join('outputs', '03_manual_parse_title', '03_01_manual_parse_rdy_for_abstract.csv'),
    sep=',',
    encoding='utf-8',
    index=False)
```

```python
## only id, title and abstract 
df_manual_parse_vote_yes_only_id_tit_abs = df_manual_parse_vote_yes[['id', 'title', 'abstract']]
df_manual_parse_vote_yes_only_id_tit_abs
```

### Export three random selections

```python
from src.main.python.at.jku.scsl.main.util.TrueRandomizer import TrueRandomizer

randomizer = TrueRandomizer(len(df_manual_parse_vote_yes_only_id_tit_abs))

for i in range(3):
    rand_ints = pd.DataFrame({'ids': randomizer.get_new_rand_ints()})
    df_rand_first = df_manual_parse_vote_yes_only_id_tit_abs[
        df_manual_parse_vote_yes_only_id_tit_abs.id.isin(rand_ints.ids)]
    df_rand_first.to_csv(os.path.join('outputs', '04_manual_parse_abstract', f'0{i + 1}_rnd_abstract.csv'), 
                         sep=',',
                         encoding='utf-8',
                         index=False)
```
