# Load Data 

```python
from src.main.python.at.jku.scsl.util.util import concat_path, get_project_root_directory
import pandas as pd

ms_df = pd.read_csv(concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results',
                                'MS_Pre_Exclusion_abstracts_All_Paper.csv'), sep=';')
ms_df = ms_df.rename(columns={'relevant?': 'relevant_ms'})

sam_df = pd.read_csv(concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results',
                                 '01_rnd_abstract_SAM.csv'))
sam_df = sam_df.rename(columns={'relevant': 'relevant_sam'})

mike_df = pd.read_csv(concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results',
                                  '02_rnd_abstract_MIKE.csv'))
mike_df = mike_df.rename(columns={'relevant': 'relevant_mike'})

mv_df = pd.read_csv(concat_path(get_project_root_directory(), 'main', 'input', '04_abstract_parsing_results',
                                '03_rnd_abstract_MV.csv'))
mv_df = mv_df.rename(columns={'relevant': 'relevant_mv'})

```

 # Sam

```python
# compare sam
merged_df_sam = pd.merge(ms_df, sam_df, on='id', suffixes=('_ms', '_sam'))

disagreements = merged_df_sam[(merged_df_sam['relevant_ms'] == 'n') & (merged_df_sam['relevant_sam'] == 'y')]
percentage_disagreement = (len(disagreements) / len(merged_df_sam)) * 100

# Reorder columns to make 'relevant_ms' and 'relevant_sam' second and third
cols = disagreements.columns.tolist()
# Move 'relevant_ms' and 'relevant_sam' to the second and third positions
# Assuming 'id' should be the first column
new_order = [cols[0], 'relevant_ms', 'relevant_sam'] + [col for col in cols if col not in ['id', 'relevant_ms', 'relevant_sam']]
disagreements = disagreements[new_order]

print(f'Percentage of Disagreements: {percentage_disagreement:.2f}%')
# Display the disagreements and the percentage
disagreements
```

 # Mike

```python
# compare mike
merged_df_mike = pd.merge(ms_df, mike_df, on='id', suffixes=('_ms', '_mike'))

disagreements = merged_df_mike[(merged_df_mike['relevant_ms'] == 'n') & (merged_df_mike['relevant_mike'] == 'y')]
percentage_disagreement = (len(disagreements) / len(merged_df_mike)) * 100


# Reorder columns to make 'relevant_ms' and 'relevant_sam' second and third
cols = disagreements.columns.tolist()
# Move 'relevant_ms' and 'relevant_sam' to the second and third positions
# Assuming 'id' should be the first column
new_order = [cols[0], 'relevant_ms', 'relevant_mike'] + [col for col in cols if col not in ['id', 'relevant_ms', 'relevant_mike']]
disagreements = disagreements[new_order]


# Display the disagreements and the percentage
print(f'Percentage of Disagreements: {percentage_disagreement:.2f}%')
disagreements
```
